Courbes des solutions du modèle SIR pour $\beta=0.5$ et $\gamma=0.33$:

```
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
plt.style.use('ggplot')

# Total population, N.
N = 7900000
# Initial number of infected and recovered individuals, I0 and R0.
I0, R0 = 10, 0
# Everyone else, S0, is susceptible to infection initially.
S0 = N - I0 - R0

# Scaled variables
s0 = S0/N
i0 = I0/N
r0 = R0/N

print(s0,i0,r0)

# Contact rate, beta, and mean recovery rate, gamma, (in 1/days).
beta, gamma = 0.5, 0.33

# A grid of time points (in days)
t = np.linspace(0, 160, 160)

# The SIR model differential equations.
def deriv(y, t, beta, gamma):
    s, i, r = y
    dsdt = -beta * s * i
    didt = beta * s * i - gamma * i
    drdt = gamma * i
    return dsdt, didt, drdt

# Initial conditions vector
y0 = s0, i0, r0
# Integrate the SIR equations over the time grid, t.
ret = odeint(deriv, y0, t, args=(beta, gamma))
s, i, r = ret.T

# Plot the data on three separate curves for s(t), i(t) and r(t)
plt.figure()
plt.plot(t, s, 'b', alpha=0.5, lw=2, label='Susceptible')
plt.plot(t, i, 'r', alpha=0.5, lw=2, label='Infected')
plt.plot(t, r, 'g', alpha=0.5, lw=2, label='Recovered')
plt.xlabel('Time (days)')
plt.ylabel('Number (rescaled)')
plt.ylim(0,1.1)
plt.legend()
plt.show()
```
